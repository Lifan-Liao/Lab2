```python
--SQL
select distinct on (first_name, last_name) first_name, last_name, title, release_year from (
 	public.actor a
 	left join public.film_actor fa
 	on a.actor_id = fa.actor_id
 	right join (SELECT * FROM public.film where rating = 'NC-17') f
    on fa.film_id = f.film_id) comb
order by last_name
limit 3
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
film_df = pd.read_csv(r'film.csv', delimiter='|', quotechar="'")
categories_df = pd.read_excel(r'categories.xlsx')
df = pd.merge(film_df, categories_df, left_on="category_id", right_on="category_id")
```

```python
# Visulization
category_counts = df.groupby("category_name")["film_id"].count()
category_counts.plot(kind='bar')
plt.title("Number of Films by Category")
plt.ylabel("Count")
plt.xlabel("Category Name")
```

```python
df.boxplot(column="rental_duration", by='category_name', showmeans=True, rot=90)
plt.title('Rental Duration Distribution by Category')
plt.xlabel('Category Name')
plt.ylabel('Rental Duration')
```
